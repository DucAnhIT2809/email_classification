```python
data = pd.read_csv('/content/drive/MyDrive/spam.csv',encoding= "ISO-8859-1")
data.shape
```

```python
data.head(5)
```

```python
df = data[['v1','v2']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(df['v2'], df['v1'], 
    test_size=0.25,random_state=8)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing import sequence
tokenizer = Tokenizer(num_words=200)
tokenizer.fit_on_texts(X_train)
X_tok_train =  tokenizer.texts_to_sequences(X_train)
X_train = sequence.pad_sequences(X_tok_train, maxlen=150)
tokenizer.fit_on_texts(X_val)
X_tok_val =  tokenizer.texts_to_sequences(X_val)
X_val = sequence.pad_sequences(X_tok_val, maxlen=150)
```

```python
y_train=y_train.map({'ham':1,'spam':0})
y_val=y_val.map({'ham':1,'spam':0})
```

```python
from keras.layers.serialization import activation

from keras.api._v2.keras import activations
from keras.layers import BatchNormalization,Dropout
from tensorflow.keras.layers import Input, Embedding, Dense, Dropout, GlobalAveragePooling1D,TextVectorization
from tensorflow.keras.models import Model, Sequential
model = keras.Sequential()
model.add(Input(shape=(150,)))
model.add(Embedding(input_dim = 200,output_dim=1,input_length=200))
model.add(GlobalAveragePooling1D())
model.add(Dense(16,activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam',
              loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train,y_train,  epochs=20,validation_data=(X_val, y_val))
```
